```python
import pandas as pd
import pandas_ta as ta
import yfinance as yf
from candlestick import Candlestick
from candlestick import CandleColor
```

```python
start_dt = '2023-03-13'
end_dt = '2023-03-24'
symbol = "TCS.NS"

df = yf.Ticker(ticker=symbol).history(period="1y", interval="1D")

df
```

```python
for row in range(1,len(df.index)):
    candle = df.iloc[row]
    df['ShootingStar'] = Candlestick(open=candle.Open, 
                                     high=candle.High, 
                                     low=candle.Low, 
                                     close=candle.Close).ShootingStar(colorFilter=CandleColor.RED)
    
    df['Hammer'] = Candlestick(open=candle.Open,
                                     high=candle.High,
                                     low=candle.Low,
                                     close=candle.Close).Hammer(colorFilter=CandleColor.GREEN)
    

```

```python
df.to_csv('candle.csv')

df
```
