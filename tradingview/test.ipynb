```python
df = pd.read_csv('crudeoil.csv')

df

# df["date"] = df["symbol"].str.split("    ").str[0]
# df
```

```python
df['EMA_9'] = ta.ema(df['close'], length=9)
df['EMA_21'] = ta.ema(df['close'], length=21)
df['EMA_55'] = ta.ema(df['close'], length=55)
```

```python
df['Supertrend'] = ta.supertrend(
    df['high'], df['low'], df['close'], length=10, multiplier=3)['SUPERT_10_3.0']

# a = []
# n = 220
# for i in range(n, len(df)):
#    a.append(df['Supertrend'][i-1])

# df['Sup_Signal'] = 'No Signal'
```

```python
df['EMA_Signal'] = ''
df['Sup_Signal'] = ''
df['Final_Signal'] = ''
n = 10
for i in range(n, len(df)):
    if df['close'][i] > df['Supertrend'][i]:
        df['Sup_Signal'][i] = 'Green'
        
    if df['close'][i] <= df['Supertrend'][i]:
        df['Sup_Signal'][i] = 'Red'
        
    if df['close'][i] > df['Supertrend'][i-1] and df['close'][i-1] <= df['Supertrend'][i-1]:
        df['Sup_Signal'][i] = 'Buy'
        
    if df['close'][i] < df['Supertrend'][i-1] and df['close'][i-1] >= df['Supertrend'][i-1]:
        df['Sup_Signal'][i] = 'Sell'
        
    if df['EMA_9'][i] > df['EMA_21'][i]:         #and df['EMA_21'][i] > df['EMA_55'][i]:
            df['EMA_Signal'][i] = 'Green'
    if df['EMA_9'][i] < df['EMA_21'][i]:
            df['EMA_Signal'][i] = 'Red'
```

```python
n = 10
for i in range(n, len(df)):
    if (df['EMA_Signal'][i] == 'Green' and (df['Sup_Signal'][i] == 'Green' or df['Sup_Signal'][i] == 'Buy')) and (df['EMA_Signal'][i-1] == 'Red' or df['Sup_Signal'][i-1] == 'Red'):
        df['Final_Signal'][i] = 'Sure Buy'
    else:
        df['Final_Signal'][i] = 'NA'

    if (df['EMA_Signal'][i] == 'Red' and (df['Sup_Signal'][i] == 'Red' or df['Sup_Signal'][i] == 'Sell')) and (df['EMA_Signal'][i-1] == 'Green' or df['Sup_Signal'][i-1] == 'Green'):
            df['Final_Signal'][i] = 'Sure Sell'
# df.to_csv('backtesting.csv')

df.tail(10)
```
