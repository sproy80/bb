```python
import yfinance as yf
import pandas as pd
import manager_db
```

```python
tickers = manager_db.get_future_stock_list()

for ticker in tickers:
    tickr = ticker.symbol + '.NS'
    print(f"Pulling data for {tickr}")
    data = yf.download(tickr, group_by="Ticker", period='2d', interval='15m')
    data['symbol'] = ticker.symbol
```
